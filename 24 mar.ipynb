```python
Mean, Median, or Mode Imputation:
Advantages:
Simple and easy to implement.
Suitable for numerical data with a small proportion of missing values.
Disadvantages:
Ignores relationships between variables.
May distort the distribution of the imputed variable, especially in the presence of outliers.
Forward Fill or Backward Fill:

Advantages:
Useful for time series data where missing values are likely to have a temporal relationship.
Disadvantages:
Assumes a constant trend between observations.
Can propagate errors if the data has sudden changes.
Interpolation:

Advantages:

Preserves the overall shape of the variable.
Suitable for time series or continuous data.
Disadvantages:

Sensitive to outliers.
Performance depends on the method chosen (linear, polynomial, spline).
Multiple Imputation:

Advantages:

Accounts for uncertainty in imputed values.
Preserves variability and relationships in the data.
Disadvantages:

More computationally intensive.
Requires assumptions about the distribution of missing data.
```

```python
import numpy as np
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
df.head()
```

```python
X = df.drop('type',axis = 1)
y= df['type']
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
pca = PCA()
X_pca = pca.fit_transform(X_scaled)
cumulative_variance = np.cumsum(pca.explained_variance_ratio_)
# Determine the number of components needed for 90% variance
num_components_90 = np.argmax(cumulative_variance >= 0.9) + 1
```
